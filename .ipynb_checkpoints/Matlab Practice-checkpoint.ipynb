# Matplotlib Practice

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Data we want to plot
x = np.arange(1,11)
x
```

```python
y = np.log(x)
y
```

#### Basic Plotting

```python
plt.plot(x,y)
```

```python
plt.barh(x,y)
```

```python
plt.plot(x,y,color="r")
plt.plot(x,y+1,color="b")
plt.plot(x,y+2,"g-+")
plt.xlabel("x values")
plt.ylabel("y=ln(x)")
```

### Subplots

```python
plt.subplot(1,2,1)
ax1 = plt.plot(x,y,color="r")
plt.subplot(1,2,2)
ax2 = plt.plot(x,y,color="b")
plt.subplot(1,2,1)
ax3 = plt.plot(x,y+1,"r*")
plt.show()
plt.subplot(2,2,3)
ax4 = plt.plot(x,x**2,"g*")
plt.show()
```

### Matplotlib Object Oriented Methods

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'b')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel("big x")
ax.set_ylabel("big y")
ax.set_title("Big Title")
ax.plot(x,x**2,'k-')
ax2 = fig.add_axes([.2,.3,.4,.4])
ax2.bar(x,y,color="r")
ax2.set_title("Mini Title")
ax2.set_xlabel("mini x")
ax2.set_ylabel("mini_y")
ax3 = fig.add_axes([.7,.1,.2,.3])
ax3.barh(y,x,color="g")
```

### subplots()

```python
fig, ax = plt.subplots(2,2)
fig.set_figheight(3)
fig.set_figwidth(3)
fig.set_dpi(200)
ax[0,0].plot(x,y)
ax[0,0].set_title("First")
ax[0,1].plot(x,x**2,color="red")
ax[0,1].set_title("Second")
ax[1,0].plot(x,np.e**x,color="blue")
ax[1,0].set_title("Third")
ax[1,1].plot(x,np.log(x),color="green")
ax[1,1].set_title("Fourth")

plt.tight_layout()
```

### Legends

```python
fig, ax = plt.subplots(2,2)
fig.set_figheight(3)
fig.set_figwidth(3)
fig.set_dpi(200)
ax[0,0].plot(x,y,label="y")
ax[0,0].plot(x,y+1,label="y+1")
ax[0,0].legend()
ax[0,0].set_title("First")
ax[0,1].plot(x,x**2,color="red")
ax[0,1].set_title("Second")
ax[1,0].plot(x,np.e**x,color="blue",linewidth=.3)
ax[1,0].set_title("Third")
ax[1,1].plot(x,np.log(x),color="green",linewidth=3)
ax[1,1].set_title("Fourth")

plt.tight_layout()
```
